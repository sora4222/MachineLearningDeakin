# EDA
Now to do my own exploration of the iris data

```python
import seaborn as sns
```

```python
fig1 = sns.FacetGrid(data=iris, hue=" class", size=10, aspect=1.2)
fig1.map(plt.scatter, 'sepal length', ' petal length').add_legend()
```

```python
fig2 = sns.FacetGrid(data=iris, hue=" class", size=10, aspect=1.2)
fig2.map(plt.scatter, 'sepal length', ' sepal width').add_legend()
```

```python
_ = sns.pairplot(iris, hue=" class", size=3, aspect=1.4)
```

## Results
The exploratory analysis, tells me that I am working with a dataset with three types of orchid. The first iris-setosa is different from the other two in that its features are much more different than the other two. Mos of the iris-setosa's features are smaller than the other two types. The other two types are harder to distinguish from each other with each of the features I have access to having a small amount of overlap even when checking pairs of features against each other. The graphs sepal width vs petal length and petal width vs sepal width show the greatest degree of seperation among the samples of each type of flower. Sepal length vs sepal width shows the least divide between the iris-versicolor and iris-virginica which are the two hardest types to seperate.
